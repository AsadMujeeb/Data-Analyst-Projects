# Exploratory Data Analysis


# Gender

```python
df.columns
```

```python
ax = sns.countplot(x = "Gender", data = df)
```

```python
ax = sns.countplot(x = "Gender", data = df)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
```

# Relation between Gender and Amount

```python
sales_gender = df.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.barplot(x = "Gender", y = "Amount", data = sales_gender)
```

Insights : From the above analysis , the purchasing power of females is greater than males

# Age

```python
df.columns
```

```python
sns.countplot(data = df, x = "Age Group", hue = "Gender")
```

```python
ax = sns.countplot(data = df, x = "Age Group", hue = "Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

# total amount by sale group

```python
sales_age = df.groupby(["Age Group"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.barplot(data = sales_age, x = "Age Group", y = "Amount")
```

Insights : Most buyers are at the age group of 26-35 females

# state

```python
df.columns
```

```python
state_orders = df.groupby(["State"], as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False).head(10)

sns.set(rc = {"figure.figsize" : (22,8)})

sns.barplot(data = state_orders, x = "State", y = "Orders")

```

# total amount and state

```python
state_amount  = df.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)

sns.set(rc = {"figure.figsize" : (22,8)})


sns.barplot(data = state_amount, x = "State", y = "Amount")
```

Insights : Most of the order & total sales/ amount are from Uttar pradesh, Maharashta and Karnashat 

# Martial Status 

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = "Marital_Status")

sns.set(rc = {"figure.figsize" : (5,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

# Marital status and gender relation

```python
sales_amount = df.groupby(["Marital_Status", "Gender"],
                          as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc = {"figure.figsize" : (6,6)})

sns.barplot(data = sales_amount, x = "Marital_Status", y = "Amount", hue = "Gender")
```

Insights : The buyer's are married (women) and they have high purchasing power

# Occupation

```python
ax  =  sns.countplot(data = df, x = "Occupation")

sns.set(rc = {"figure.figsize" : (22,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

# Occupation and Amount Relation

```python
sales_amount = df.groupby(['Occupation'], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.barplot(data = sales_amount, x = "Occupation", y = "Amount")
```

Insights : The most purchasing power is in IT Sector, Healthcare and Aviation etc

# Product Category

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = "Product_Category")

sns.set(rc = {"figure.figsize" : (20,9)})

for bars in ax.containers:
    ax.bar_label(bars)
```

# Amount and Product Category

```python
sales_amount = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False).head(7)

sns.barplot(data = sales_amount, x = "Product_Category", y = "Amount")
```

Insights : The product category Food, clothing and electonics are most selling products category
